# Explore San Francisco Meetups

## Prerequesites:
Add a Meetup API Key to your environment.
If you need a Meetup API Key you can fetch it from [here](https://secure.meetup.com/meetup_api/key/)

````python
import datetime
import matplotlib.pyplot as plt
import math
import meetup.api
import os
import pandas as pd
import pprint
import requests

%matplotlib inline
````

## Double check your environment

Nothing works without **MEETUP_API_KEY**.

````python
assert 'MEETUP_API_KEY' in os.environ, "You need a MEETUP_API_KEY in your environment please look at the README for instructions."
client = meetup.api.Client()
````

## Categories

There are multiple categories of groups in Meetup, let's use Python's meetup.api to [GetCategories](https://meetup-api.readthedocs.io/en/latest/meetup_api.html#meetup.api.meetup.api.Client.GetCategories).

````python
categories = client.GetCategories()
````

### Exploring the response object

By looking at the **meta** member of the response, we can see that there are 33 different categories.

### What are the attributes of the response object?
We can find out by using __dict__ to get the attribute dictionary.

````python
import json
categories.__dict__
````

### First, let's look at the meta-data

````python
def parse_meta(response):
    return pd.DataFrame.from_dict(response.meta, orient='index')

cat_meta_df = parse_meta(categories)
cat_meta_df
````

### Secondly, let's take a look at the categories themselves

````python
def parse_results(response):
    return pd.DataFrame.from_dict(response.results)

cats_df = parse_results(categories)
cats_df
````

### So, if we want to work with a particular category
In our case 'Tech'...

````python
tech_df = cats_df.loc[cats_df['name'] == 'Tech']
tech_df
````

### Let's go ahead and store the ID number for later reference

````python
tech_category_id = tech_df['id'].values[0]
tech_category_id
````

## Explore Cities
### Now let's look at cities in the United States named San Francisco

````python
cities_resp = client.GetCities(country='United States', query='San Francisco')
````

````python
cities_meta_df = parse_meta(cities_resp)
cities_meta_df
````

### Hmm. Why are there 10 cities?

````python
cities_df = parse_results(cities_resp)
cities_df
````

### Let's filter the dataframe to just cities in California, US

````python
single_city_df = cities_df.loc[
    (cities_df['state'] == 'CA') & 
    (cities_df['country'] == 'us')]

single_city_df
````

### And store the latitude and longitude for later

````python
latitude, longitude = single_city_df['lat'].values[0], single_city_df['lon'].values[0]
latitude, longitude
````

## Explore Groups

````python
group_resp = client.GetGroups(
    category_id=tech_category_id, lat=latitude, lon=longitude, offset=0)
````

````python
sf_group_meta = parse_meta(group_resp)
sf_group_meta
````

````python
first_page_df = parse_results(group_resp)
first_page_df
````

### That's too many fields
I think I'll limit my columns to just what I care about

````python
first_page_df = first_page_df[['id', 'name', 'members', 'rating', 'join_mode', 'urlname']]
first_page_df.head()
````

## Getting all Tech Meetup Groups

````python
# searching for Vancouver, Canada meetups
meetup_groups = []

group_resp = client.GetGroups(category_id = 34, city='San Francisco', state='ca', country='us', offset=0)
total_count = group_resp.meta['total_count']
num_groups = len(group_resp.results)
meetup_groups.extend(group_resp.results)

page_num = math.ceil(total_count / num_groups)
page_num
````

````python

for i in range(1, page_num):
    print(f"{i} - Getting groups")
    group_resp = client.GetGroups(category_id = 34, city='San Francisco', state='ca', country='us', offset=i)
    meetup_groups.extend(group_resp.results)
````

````python
len(meetup_groups)
````

---

## Get Group Events

````python
first_ten = meetup_groups[:10]
ids = [str(group['id']) for group in first_ten]
ids
````

````python
','.join(ids)
````

````python
# get events
events_resp = client.GetEvents(group_id=','.join(ids), status='past')
````

````python
dir(events_resp)
````

````python
events_resp.meta
````

````python
pprint.pprint(events_resp.results[0])
````

````python
events_result = events_resp.results
````

````python
unique_status = set()

for event in events_result:
    unique_status.add(event['status'])

print(unique_status)
````

````python
for event in events_result[:10]:
    event_name = event['name']
    group_name = event['group']['name']
    event_time_seconds = event['time'] / 1000.0
    event_time = datetime.datetime.fromtimestamp(event_time_seconds).strftime('%a %Y-%m-%d')
    print(f"{event_time}: {group_name:<53}: {event_name}")
````
